## creating a dataframe from the text files containing data of dialoges,speaker,Season,Episode,TV Show

```python
def dialogues(path):
    
    f=open(path,'r')

    l=[]
    li=f.readlines()
    for i,j in enumerate(li):
        if ':' in j:
            sp=j[:j.index(':')]
            new=j[j.index(':')+1:]
        else:
            sp=np.nan
            new=j
        a=(new[:-1],sp)
        l.append(a)
    s=pd.DataFrame(l)
    
    s.columns=['Text','Speaker']
    s.set_index('Text',inplace=True)
    
    return s
```

```python
import os 
    

df=pd.DataFrame(columns=['Text','Speaker','Episode','Season','Show'])
shows=os.listdir('TV Shows')

for show in shows:
    seasons=os.listdir("TV Shows/" + show)
    for season in seasons:
        episodes = os.listdir("TV Shows/" + show+ "/" + season)
        for episode in episodes:
            text1=dialogues('TV Shows/'+ show  +'/' + season +'/'+ episode)
            
            show1=pd.DataFrame([show] * text1.shape[0])
            show1.index=text1.index
            season1=pd.DataFrame([season] * text1.shape[0])
            season1.index=text1.index
            episode1=pd.DataFrame([episode[:-4]] * text1.shape[0])
            episode1.index=text1.index
            temp_df=pd.concat([text1,episode1,season1,show1],axis=1)
            #temp_df.set_index('Text',inplace=True)
            temp_df.columns=['Speaker','Episode','Season','Show']
            df=pd.concat([df,temp_df],axis=0)
            #df.set_index('Text')
            
            
df.drop('Text',axis=1,inplace=True)

```

```python
def script(path):
    f=open(path,'r')

    l=[]
    s=f.readlines()
    for i in s:
        a=i[:-2]
        l.append(a)
    s='. '.join(l)
    s=pd.DataFrame([s])
    s.columns=['Text']
    #s.set_index('Text',inplace=True)
    return s
```

```python
import os 
    

df=pd.DataFrame(columns=['Text','Episode','Season','Show'])
shows=os.listdir('TV Shows')

for show in shows:
    seasons=os.listdir("TV Shows/" + show)
    for season in seasons:
        episodes = os.listdir("TV Shows/" + show+ "/" + season)
        for episode in episodes:
            text1=script('TV Shows/'+ show  +'/' + season +'/'+ episode)
            
            show1=pd.DataFrame([show])
            show1.index=text1.index
            season1=pd.DataFrame([season])
            season1.index=text1.index
            episode1=pd.DataFrame([episode[:-4]])
            episode1.index=text1.index
            temp_df=pd.concat([text1,episode1,season1,show1],axis=1)
            #temp_df.set_index('Text',inplace=True)
            temp_df.columns=['Speaker','Episode','Season','Show']
            df=pd.concat([df,temp_df],axis=0)
            #df.set_index('Text')
            
            
df.drop(['Text'],axis=1,inplace=True)
df.columns=['Episode','Season','Show','Text']
df.reset_index(drop=True,inplace=True)
```
